Create a 2D numpy array A and a 1D numpy array
$$ A = \begin{bmatrix}

x = \begin{bmatrix}
1 \\ -1 \\ 2

and compute $Ax$.

```python
#CODE HERE

x = np.array([1, -1, 2])
print(x)

A = np.array([[1, 2, 3], [4, 5, 6]])
print(A)

A@x
```

Construct a function that multipies a matrix times a vector

```python
def matrix_vector_mult(A,x):
    """
    This function returns the product of the matrix A and vector x.
    The product is computed by looping over columns only.
    There is no check that the the number of columns of A is equal to the number of elements of x
    """
    nrows, ncolumns = A.shape
    #CODE HERE
    if ncolumns != x.shape[0]:
        raise Exception('This multipication is not possible.')
    b = np.zeros(nrows) 
    for ii in range(0,ncolumns):
        b += x[ii] * A[:,ii]
    return b
matrix_vector_mult(A,x)
```

### 3.2 Compute the climatology and anomaly

Compute the climatology of Nino 3.4 in 3 different ways and get 3 different answers. 

1. Which one is correct? 
2. Explain the errors Fix the errors in the other two.

1. Method 3 (xarray groupby) is correct! 
2. In method 2. we added an order='F'in order to sort the whole data series in a reasonable way that each column stands per year, and each row stands per month. & also fixed the climatology subtraction to climatology2. 
2. In method 1, we changed the in range to first element of 0. and the second line of months we switched 12 and nmonths; so that nmotnhs is equal to stop and 12 refeltcs the steps. and finally we changed last line so that it can be subtracted total months. 

```python
# 41 years
nmonths = n34.size
nmonths/12
#climatogology= avg of month of months - jenn note 
#range function in method 1 
#method 2, rows and columns 
```

Method 1 (loops)

```python
climatology = np.full(12,np.nan)
anomaly = np.zeros_like(n34)
```

```python
for imonth in range(0,12):
    months = range(imonth, nmonths, 12)
    climatology[imonth] = n34[months].mean()
    anomaly[months] = n34[months] - climatology[imonth]
```

Method 2 (reshape)

```python
# reshape the data so that each of the 41 columns contains a year and then average over columns (years) 
climatology2 = n34.values.reshape(12,-1, order='F').mean(axis=1)
anomaly2 = n34.values.reshape(12,-1, order='F').T - climatology2
anomaly2.resize(nmonths)
```

Method 3 (xarray groupby)  
http://xarray.pydata.org/en/stable/groupby.html

```python
climatology3 = n34.groupby("T.month").mean()
anomaly3 = n34.groupby("T.month") - climatology3
```

```python
# plot them
plt.plot(climatology,label = 'loops')
plt.plot(climatology2,label = 'reshape')
plt.plot(climatology3, label = 'groupby')
plt.legend()
plt.xticks(np.arange(0,12), m_str)
plt.title('Nino 3.4 climatology');
```

```python
# plot them
plt.plot(anomaly,label = 'loops')
plt.plot(anomaly2,label = 'reshape')
plt.plot(anomaly3, label = 'groupby')
plt.legend()
plt.title('Nino 3.4');
```
